### Metric Learning

[A good description](https://towardsdatascience.com/how-to-use-metric-learning-embedding-is-all-you-need-f26e01597375)
[Where it is used](http://contrib.scikit-learn.org/metric-learn/introduction.html)

Metric Learning is essentially training the encoder to represent the image in a good feature space before training the fully connected layers. So for images we use a forward pass of a model to represent the image in a feature space (embeddings) and utilise a unique loss function that backpropogates through the encoder in order to generate good embeddings.
