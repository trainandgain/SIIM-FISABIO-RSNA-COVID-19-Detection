# Inference

An investigation into the previous two iterations.

```python
# imports
import torch
import json
import os
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from tqdm import tqdm
```

```python
# paths 
FIVE_PATH = '../models/iteration_5/FasterRCNNDetector/'
SIX_PATH = '../models/iteration_6/FasterRCNNDetector/'
```

```python
# load logging five
with open(os.path.join(FIVE_PATH, 'FasterRCNNDetector.json')) as file:
    five_data = file.read()
    file.close()
    
five_obj = json.loads(five_data)
print(five_obj.keys())
```

```python
with open(os.path.join(SIX_PATH, 'FasterRCNNDetector.json')) as file:
    six_data = file.read()
    file.close()
    
six_obj = json.loads(six_data)
print(six_obj.keys())
```

```python
five_models = []
six_models = []

for i in tqdm(range(five_obj['NB_EPOCHS'])):
    model = torch.load(os.path.join(FIVE_PATH, 'epoch_{}.pth'.format(i)), 
               map_location=torch.device('cpu'))
    five_models.append(model)

for i in tqdm(range(six_obj['NB_EPOCHS'])):
    model = torch.load(os.path.join(SIX_PATH, 'epoch_{}.pth'.format(i)), 
               map_location=torch.device('cpu'))
    six_models.append(model)
```

## Helper Functions

```python
def print_setup(obj):
    for k, v in five_obj.items():
        if k == 'data':
            continue
        print(f'{k}: {v}')
```

```python
def model_epochs_dataframe(model_list):
    data = {
        # model
        'epoch': [],
        'loss': [],
        'image_precision': [],
        # optmiser
        'lr': [],
        'momentum': [],
        'dampening': [],
        'weight_decay': [],
        # scheduler
        'step_size': [],
        'step_count': [],
        'gamma': [],  
    }

    for i in range(len(model_list)):
        # temporary storage
        temp = model_list[i]
        temp_optimiser = temp['optimizer_state_dict']['param_groups'][0]
        temp_scheduler = temp['scheduler_state_dict']
        # append data
        data['epoch'].append(temp['epoch'])
        data['loss'].append(temp['loss'])
        data['image_precision'].append(temp['image_precision'])
        data['lr'].append(temp_optimiser['lr'])
        data['momentum'].append(temp_optimiser['momentum'])
        data['dampening'].append(temp_optimiser['dampening'])
        data['weight_decay'].append(temp_optimiser['weight_decay'])
        data['step_size'].append(temp_scheduler['step_size'])
        data['step_count'].append(temp_scheduler['_step_count'])
        data['gamma'].append(temp_scheduler['gamma'])

    return(pd.DataFrame(data))
```

## Model Over Epochs

We can investigate the model changes over epochs.

```python
print('Keys for loaded model: {}'.format(five_models[0].keys()))
print('*'*50)
print('Keys for optimiser: {}'.format(five_models[0]['optimizer_state_dict'].keys()))
print('*'*50)
print('Keys for scheduler" {}'.format(five_models[0]['scheduler_state_dict'].keys()))
```

```python
# load df
five_df = model_epochs_dataframe(five_models)
five_df.head()
```

```python
# init plot
sns.set(rc={"figure.figsize":(20, 10)})
sns.pairplot(data=five_df, corner=True)
plt.show()
```

## Learning Rate

```python
fig, axs = plt.subplots(2, 2)
# individual
sns.lineplot(data=five_df, x='epoch', y='lr', ax=axs[0, 0]).set(title='Learning Rate over Epochs')
sns.lineplot(data=five_df, x='epoch', y='loss', ax=axs[0, 1]).set(title='Loss over Epochs')
sns.lineplot(data=five_df, x='epoch', y='image_precision', ax=axs[1, 0]).set(title='Precision Rate over Epochs')
# combined
sns.lineplot(data=five_df, x='epoch', y='loss', ax=axs[1, 1], label='loss')
sns.lineplot(data=five_df, x='epoch', y='image_precision', ax=axs[1, 1], label='precision')
sns.lineplot(data=five_df, x='epoch', y='lr', ax=axs[1, 1], label='lr').set(title='Comparison')

plt.show()
```

```python
sns.lineplot(data=five_df, x='epoch', y='image_precision', 
             hue='lr', palette='Dark2').set(title='Precision Rate over Epochs')
plt.show()
```

From the comparitive plots it looks like the learning rate is too low for the later epochs and therfore we converge. 

The first 10 epochs have the desired shape we are looking for. 

From first glance it appears that the initial learning rate (0.005) is too low and the gamma (0.1) is too low.

Perhaps starting with 0.1 and a gamma of 0.2 would suffice.
